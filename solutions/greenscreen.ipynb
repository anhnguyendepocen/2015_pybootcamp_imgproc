```python
from __future__ import division, print_function

%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

from skimage import io, img_as_float
```

```python
people = img_as_float(io.imread('../images/greenscreen.jpg'))
forest = img_as_float(io.imread('../images/forest.jpg'))
```

```python
people_red = people[..., 0]
people_green = people[..., 1]
people_blue = people[..., 2]
```

```python
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(10, 8))
ax0.imshow(people_red, cmap='gray')
ax1.imshow(people_green, cmap='gray')
ax2.imshow(people_blue, cmap='gray');
```

```python
plt.hist(people_red.ravel(), bins=255);
```

```python
plt.hist(people_green.ravel(), bins=255);
```

```python
mask = (people_red > 0.02) & (people_green < 0.8) | (people_blue > 0.6)
plt.imshow(mask, cmap='gray');
```

```python
new_forest = forest.copy()
new_forest[np.where(mask)] = people[mask]
```

```python
plt.imshow(new_forest);
```
